```python
import pandas as pd
import numpy as np
import seaborn as sb
from sklearn.preprocessing import StandardScaler
import os
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from sklearn import preprocessing
d = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv", delimiter=';')
d.columns = ['fixed_acidity','volatile_acidity','citric_acid','residual_sugar','chlorides','free_sulfur_dioxide','total_sulfur_dioxide','density','pH','sulphates','alcohol','quality']
d = d.replace(' ?',np.NaN)
data = d.drop(["quality"],axis=1)
trueCluster = d.drop(['fixed_acidity','volatile_acidity','citric_acid','residual_sugar','chlorides','free_sulfur_dioxide','total_sulfur_dioxide','density','pH','sulphates','alcohol'],axis=1)
print('Number of instances = %d' % (data.shape[0]))
print('Number of attributes = %d' % (data.shape[1]))
print('column has = ', (data.columns))
```

```python
trueClusters=trueCluster-3
```

```python
dups = data.duplicated()
print('Number of duplicate rows = %d' % (dups.sum()))
print('Number of missing values:')
for col in data.columns:
    print('\t%s: %d' % (col,data[col].isna().sum()))
data.describe()
df=((data-data.min())/(data.max()-data.min()))
scaled_df = df.values[:,0:11]
```

```python
def EuclideanDistance(df,center):
    dist = np.zeros((df.shape[0],center.shape[0]))
    for c in range(center.shape[0]):
        indice=0
        for i in df: 
            #print(indice, ' at ',i)
            dist[indice,c] = np.sqrt(pow((i[0]-center[c][0]),2)+pow((i[1]-center[c][1]),2)+pow((i[2]-center[c][2]),2)+pow((i[3]-center[c][3]),2)+pow((i[4]-center[c][4]),2)+pow((i[5]-center[c][5]),2)+pow((i[6]-center[c][6]),2)+pow((i[7]-center[c][7]),2)+pow((i[8]-center[c][8]),2)+pow((i[9]-center[c][9]),2)+pow((i[10]-center[c][10]),2))
            #print(dist)
            indice=indice+1        
    return dist
```

```python
################################################################################################################################
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=1, random_state=0).fit(scaled_df)
labels = kmeans.predict(scaled_df)
centroids = kmeans.cluster_centers_
#print(centroids)

Dist = EuclideanDistance(scaled_df,centroids)

TotalSSE=0

ind = 0
for i in Dist:
    mini = min(i)
    TotalSSE += pow(mini,2)  
    ind+=1
print('Total SSE at K = 1:',TotalSSE)
```

```python
Sum_of_squared_distances = []
K = range(1,15)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(scaled_df)
    Sum_of_squared_distances.append(km.inertia_)
```

```python
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```
