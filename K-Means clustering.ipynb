```python
#Author: Han Han
#Date: 3/28/2019

import pandas as pd
import numpy as np
import seaborn as sb
data = pd.read_csv(r"C:\Users\mrj\Desktop\TwoDimHard.csv")
data.columns = ['ID','x1','x2','TrueCluster']
print('Number of instances = %d' % (data.shape[0]))
print('Number of attributes = %d' % (data.shape[1]))
print('column has = ', (data.columns))
```

```python
def Kmeans(k,data):

    df = data.values[:,1:3]
    trueCluster = data.values[:,3]
    trueCenter=np.zeros((4,2))
    for i in range(4): 
        n=0
        points=0
        for d in range(400):
            if trueCluster[d]==(i+1):
                points +=df[d]
                n+=1
        trueCenter[i]=points/n;
    #print(trueCenter)
################################################################################################################################
    from matplotlib import pyplot as plt
################################################################################################################################
    mean = np.mean(df, axis = 0)
    std = np.std(df,axis = 0)
    c = df.shape[1]
    center = np.random.randn(k,2)*std+mean
    #print("center",center)
################################################################################################################################
    colors =['blue','yellow','red','green']
    for i in range(df.shape[0]):
        plt.scatter(df[i,0], df[i,1], s = 7, color = colors[int(trueCluster[i])-1])
        plt.scatter(center[:,0],center[:,1],marker='*',c='black',s = 150)
    plt.title('Random generated Clusters')    

################################################################################################################################
    def EuclideanDistance(df,center):
        dist = np.zeros((df.shape[0],k))
        for c in range(k):
            d=0
            indice=0
            for i in df:    
                dist[indice,c] = np.sqrt(pow((i[0]-center[c][0]),2)+pow((i[1]-center[c][1]),2))
                #print(dist)
                indice=indice+1        
        return dist

################################################################################################################################
    import copy
    centers_old = np.zeros(center.shape) 
    clusters = np.zeros(400)
    distances = np.zeros((df.shape[0],k))

    error=0
    difference = 0
    difference = center - centers_old 
    for i in difference:
        error += i[0]+i[1]

    while error != 0:
        SSE=0
        SSECluster1=0
        SSECluster2=0
        SSECluster3=0
        SSECluster4=0

        distances = EuclideanDistance(df,center)
        ind = 0
        for i in distances:
            mini = min(i)
            SSE += pow(mini,2)
            newCluster = np.where(i==mini)[0]
            if newCluster ==0:
                SSECluster1+=pow(mini,2)
            if newCluster ==1:
                SSECluster2+=pow(mini,2)
            if newCluster ==2:
                SSECluster3+=pow(mini,2)
            if newCluster ==3:
                SSECluster4+=pow(mini,2)
            clusters[ind]=newCluster
            ind+=1
        centers_old = copy.deepcopy(center)
        for i in range(k):
            n=0
            points=0
            for d in range(400):
                if clusters[d]==i:
                    points +=df[d]
                    n+=1
            if n ==0:
                n=1
            center[i]=points/n;
        error=0
        difference = center - centers_old 
        for i in difference:
            error += i[0]+i[1]
    print('Total SSE at K = ',k,' :',SSE)
    print('Cluster 1 SSE at K = ',k,' :',SSECluster1)
    print('Cluster 2 SSE at K = ',k,' :',SSECluster2)
    print('Cluster 3 SSE at K = ',k,' :',SSECluster3)
    print('Cluster 4 SSE at K = ',k,' :',SSECluster4)

################################################################################################################################
    plt.figure(figsize=(15,15))
    fig = plt.subplot(2,2,1)
    for i in range(df.shape[0]):
        plt.scatter(df[i,0], df[i,1], s = 7, color = colors[int(trueCluster[i])-1])
    plt.scatter(trueCenter[:,0], trueCenter[:,1], marker='*', c='black', s=150)
    plt.title('True Clusters')    
    plt.subplot(2,2,2)
    colors =['blue','yellow','red','green']
    for i in range(df.shape[0]):
        plt.scatter(df[i,0], df[i,1], s = 7, color = colors[int(clusters[i])-1])
        plt.scatter(center[:,0],center[:,1],marker='*',c='black',s = 150)
    plt.title('K Means generated Clusters with')   
    
    from sklearn.metrics import confusion_matrix
    trueCluster = data.values[:,3]-1
    
    print("\nCross Tabulation Matrix at K =",k,)
    print("Row:Predicted Cluster, Column:True Cluster")
    print(pd.DataFrame(confusion_matrix(clusters,trueCluster,labels=[0,1,2,3])))
    
    
    
    
    return SSE
```

```python
SSE4=Kmeans(4,data)
```

```python
SSB3 = SSE1- SSE3
print('SSB at K = 3:',SSB3)
```
